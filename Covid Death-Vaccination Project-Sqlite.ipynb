```python
#Allows access to SQLite
import sqlite3 as sql
# Packages for data manipulation
import numpy as np
import pandas as pd
import plotly.express as px
import os
```

```python
# Load up csv files
covid_deaths=pd.read_csv('/kaggle/input/covid-deathvacc/Covid Deaths.csv')
covid_vacc=pd.read_csv('/kaggle/input/covid-deathvacc/Covid Vacc.csv')
```

```python
# Create database connection
connection = sql.connect('covid.db')
```

```python
# Insert data into database
covid_deaths.to_sql('covid_deaths', connection, if_exists='replace', index=False)
covid_vacc.to_sql('covid_vacc', connection, if_exists='replace', index=False)
```

```python
# Looking at total cases vs total deaths

df3 = pd.read_sql("""
    SELECT location, date, total_cases,total_deaths, (total_deaths/total_cases)*100 As death_percentage
    From covid_deaths
    Where continent is not null
    order by 1,2; """, connection
)
df3
```

```python
# Look at total cases vs total death specific to the United States
# Shows what percentage of total covid cases that resulted in death 

df4 = pd.read_sql("""
    SELECT location, date, total_cases,total_deaths, (total_deaths/total_cases)*100 As death_percentage
    From covid_deaths
    Where location = 'United States'And continent is not null
    order by 1,2; """, connection
)
df4
```

```python
# Looking at total cases vs population specific to the United States 
# Shows what percentage of the population got Covid 

df5 = pd.read_sql("""
    SELECT location, date, total_cases,population, (total_cases/population)*100 As percent_of_pop_infected
    From covid_deaths
    Where location = 'United States' and continent is not null
    order by 1,2; """, connection
)
df5
```

```python
# Looking at which countries have the highest infection rate compared to population

df6 = pd.read_sql("""
    SELECT location, population, MAX(total_cases) As highest_infection_count, MAX((total_cases/population))*100 As percent_of_pop_infected
    From covid_deaths
    Where continent is not null
    Group by location, population
    order by percent_of_pop_infected desc; """, connection
)
df6
```

```python
# Looking at countries with highest death count per population 

df7 = pd.read_sql("""
   SELECT location, population, Max(total_deaths) as total_death_count
   From covid_deaths
   Where continent is not null
   Group by location, population
   order by total_death_count desc; """, connection
)
df7
```

```python
# Highest deaths broken down by continent 

df8 = pd.read_sql("""
   SELECT continent, Max(total_deaths) as total_death_count
   From covid_deaths
   Where continent is not null
   Group by continent
   order by total_death_count desc; """, connection
)
df8
```

```python
# Total deaths by continent

df15 = pd.read_sql("""
    Select location, Sum(new_deaths) as TotalDeathCount
    From covid_deaths
    Where continent is null And location not IN('World', 'European Union', 'International')
    Group by location
    Order By TotalDeathCount desc """, connection
)
df15
```

```python
# Highest deaths broken down by location 

df9 = pd.read_sql("""
   SELECT location, Max(total_deaths) as total_death_count
   From covid_deaths
   Where continent is null
   Group by location
   order by total_death_count desc; """, connection
)
df9
```

```python
# Looking at Global Numbers by Date

df10 = pd.read_sql("""
   SELECT date, SUM(new_cases) as total_new_cases, SUM(new_deaths) as total_new_deaths, (SUM(new_deaths)/SUM(new_cases))*100 As death_percentage
   From covid_deaths
   Where continent is not null
   Group by date
   order by 1,2; """, connection
)
df10
```

```python
# Global Numbers by Overall

df11 = pd.read_sql("""
   SELECT SUM(new_cases) as total_new_cases, SUM(new_deaths) as total_new_deaths, (SUM(new_deaths)/SUM(new_cases))*100 As death_percentage
   From covid_deaths
   Where continent is not null
   order by 1,2; """, connection
)
df11
```

```python
# Join Covid Deaths and Vaccination table 

df12 = pd.read_sql("""
   Select *
   From covid_deaths death
   Join covid_vacc vacc
   On vacc.location = death.location AND vacc.date = death.date; """, connection
)
df12

```

```python
# Looking at total population vs new vaccinations per day

df13 = pd.read_sql("""
   Select death.continent, death.location, death.date, death.population, vacc.new_vaccinations, SUM(vacc.new_vaccinations) OVER (PARTITION BY death.location order by death.location, death.date) as Rolling_People_Vacc
   From covid_deaths death
   Join covid_vacc vacc
   On vacc.location = death.location AND vacc.date = death.date
   Where death.continent is not null
   order by 2,3; """, connection
)
df13
```

```python
# Using CTE (Common Table Expression) 
# Percentage of the population that is vaccinated 

df14 = pd.read_sql("""
     With PopvsVacc
  As
  (
  Select death.continent, death.location, death.date, death.population, vacc.new_vaccinations, SUM(vacc.new_vaccinations) OVER (PARTITION BY death.location order by death.location, death.date) as Rolling_People_Vacc
  From covid_deaths death
  Join covid_vacc vacc
  On vacc.location = death.location AND vacc.date = death.date
  Where death.continent is not null
  )
  Select *, (Rolling_People_Vacc/population)*100
  From PopvsVacc; """, connection
)
df14
```
